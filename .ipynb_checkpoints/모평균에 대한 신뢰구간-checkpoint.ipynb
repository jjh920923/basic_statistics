```python
def confidence_interval(X, percent, std, n):
    if percent == 90:
        left = X - 1.65 * (std / math.sqrt(n))
        right = X + 1.65 * (std / math.sqrt(n))
    elif percent == 95:
        left = X - 1.96 * (std / math.sqrt(n))
        right = X + 1.96 * (std / math.sqrt(n))
    elif percent == 99:
        left = X - 2.58 * (std / math.sqrt(n))
        right = X + 2.58 * (std / math.sqrt(n))
    else:
        print("90%, 95%, 99% 중 하나의 신뢰구간을 고르세요.")
    print("신뢰구간 : " + str(format(left, ".2f")) + " < μ < " + str(format(right, ".2f")))
```

```python
confidence_interval(54, 95, 6, 50)
```

```python
confidence_interval(5.6, 99, 0.8, 30)
```

```python
# 모평균 구간 추정을 위해 필요한 최소 표본크기에 대한 공식
# error는 허용 오차, percent는 신뢰도, std는 표준편차
def sample_size(error, percent, std):
    if percent == 90:
        n = math.ceil(math.pow(((1.65*std) / error), 2))
    elif percent == 95:
        n = math.ceil(math.pow(((1.96*std) / error), 2))
    elif percent == 99:
        n = math.ceil(math.pow(((2.58*std) / error), 2))
    else:
        print("90%, 95%, 99% 중 하나의 신뢰구간을 고르세요.")
    print("적어도 " + str(n) + "개의 표본이 필요합니다.")
```

```python
sample_size(2, 99, 4.38)
```
