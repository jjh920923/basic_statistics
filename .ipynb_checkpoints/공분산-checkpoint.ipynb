```python
# 공분산(covariance)은 2개의 확률변수의 상관정도를 나타내는 값이다.
# 공분산은 상관관계의 상승 혹은 하강하는 경향을 이해할 수 있으나 2개 변수의 단위의
# 크기에 따라 값이 달라지므로 상관분석을 통해 정도를 파악하기에는 부적절하다.
```

```python
# 공분산 구현해보기
a = np.random.rand(3, 3)
b = np.random.rand(3, 3)
print("행렬 a\n", a)
print("---------------------------------------")
print("행렬 b\n", b)
print("---------------------------------------")
print("a의 공분산 : \n", np.cov(a))
print("---------------------------------------")
print("b의 공분산 : \n", np.cov(b))
print("---------------------------------------")
print("a와 b의 공분산 : \n", np.cov(a, b))
```
