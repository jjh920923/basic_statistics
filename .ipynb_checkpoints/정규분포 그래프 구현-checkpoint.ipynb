```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
# 입력방법 예시
# P(-1 < z < 2) 일 경우 normal_distribution("(-1 < z) & (z < 2)")
# P(z < 2)일 경우 normal_distribution("-1 < z") or normal_distribution("z < 2")
# P(2 < z) or P(z < -1)일 경우 normal_distribution("(z < -1) | (2 < z)")
```

```python
def normal_distribution(prob_range):
    if "&" in prob_range:
        z = np.linspace(-3, 3, 101)
        y = norm(0, 1).pdf(z)
        prob_range_list = [float(i) for i in re.findall("[-+]?\d*\.\d+|[-+]?\d+", prob_range)]
        area = norm(0, 1).cdf(prob_range_list[1]) - norm(0, 1).cdf(prob_range_list[0])
        plt.figure(figsize=(10, 6))
        plt.plot(z, y, color="blue")
        plt.axhline(0, color='black')
        plt.axvline(prob_range_list[0], linestyle='--', color='blue', alpha=.7)
        plt.axvline(prob_range_list[1], linestyle='--', color='blue', alpha=.7)
        plt.fill_between(z, 0, y, where=eval(prob_range), facecolor='skyblue', alpha=.8,)
        plt.text(0, .2, 'area=' + str(round(area, 4)), fontsize=14, horizontalalignment='center')
        plt.xlabel("z")                  
        plt.ylabel("y")                 
        plt.grid()                          
        plt.title("Standard Normal Distribution " + str(prob_range_list))     
        plt.legend(["N(0, 1)"])
        plt.show()

    elif "|" in prob_range:
        z = np.linspace(-3, 3, 101)
        y = norm(0, 1).pdf(z)
        prob_range_list = [float(i) for i in re.findall("[-+]?\d*\.\d+|[-+]?\d+", prob_range)]
        area = norm(0, 1).cdf(prob_range_list[0]) + (1 - norm(0, 1).cdf(prob_range_list[1]))
        plt.figure(figsize=(10, 6))
        plt.plot(z, y, color="blue")
        plt.axhline(0, color='black')
        plt.axvline(prob_range_list[0], linestyle='--', color='blue', alpha=.7)
        plt.axvline(prob_range_list[1], linestyle='--', color='blue', alpha=.7)
        plt.fill_between(z, 0, y, where=eval(prob_range), facecolor='skyblue', alpha=.8,)
        plt.text(0, .2, 'area=' + str(round(area, 4)), fontsize=14, horizontalalignment='center')
        plt.xlabel("z")                  
        plt.ylabel("y")                 
        plt.grid()                          
        plt.title("Standard Normal Distribution " + "(-∞, " + str(prob_range_list[0]) + "] & "
                  + "[" + str(prob_range_list[1]) + ", ∞)")     
        plt.legend(["N(0, 1)"]) 
        plt.show()
    
    else:
        z = np.linspace(-3, 3, 101)
        y = norm(0, 1).pdf(z)
        prob_range_list = [float(i) for i in re.findall("[-+]?\d*\.\d+|[-+]?\d+", prob_range)]
        area = float(1 - norm(0, 1).cdf(prob_range_list))
        plt.figure(figsize=(10, 6))
        plt.plot(z, y, color="blue")
        plt.axhline(0, color='black')
        plt.axvline(prob_range_list, linestyle='--', color='blue', alpha=.7)
        plt.fill_between(z, 0, y, where=eval(prob_range), facecolor='skyblue', alpha=.8,)
        plt.text(0, .2, 'area=' + str(round(area, 4)), fontsize=14, horizontalalignment='center')
        plt.xlabel("z")                  
        plt.ylabel("y")                 
        plt.grid()
        if "<" in prob_range:                      
            plt.title("Standard Normal Distribution " + "[" + str(prob_range_list[0]) + ", ∞)")
        else:
            plt.title("Standard Normal Distribution " + "(-∞, " + str(prob_range_list[0]) + "])")
        plt.legend(["N(0, 1)"]) 
        plt.show()
```

```python
normal_distribution("(z < -0.44) | (1.92 < z)")
```

```python
normal_distribution("(0 < z) & (z < 1.32)")
```

```python
normal_distribution("-1 < z")
```
