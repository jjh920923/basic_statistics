```python
# 상관계수는 데이터들의 관계가 어떻게(음, 양, 무상관) 이루어져 있는지 나타내는 지표이다.
# 대표적으로 상관계수에는 피어슨 상관계수와 스피어만 상관계수가 존재하는데
# 자주 사용하는 피어슨 상관계수에 대해서 알아보겠다.

# 피어슨 상관계수는 등간척도나 비례척도의 데이터에서 사용된다.
# 피어슨 상관계수(pearson correlation coefficient)는 두 변수 X와 Y간의 선형 상관관계를
# 계량화한 수치이다. 피어슨 상관계수는 코시-슈바르츠 부등식에 의해 +1과 -1 사이의 값을
# 가지며, 1에 가까울 수록 양의 선형관계 0에 가까울 수록 상관관계가 없음, -1에 가까울 수록
# 음의 상관관계를 의미한다.
# 피어슨 상관계수의 공식은 두 변수의 공분산을 표준편차의 곱으로 나눈 값이다.
```
