```python
# 회귀분석을 계산하기 앞서 기울기(slope)와 절편(intercept)을 구하기 위해서 최소제곱법에 대해서 알아야한다. 
# 최소제곱법은 실제값과 추정값(기댓값)의 차이의 제곱의 합을 최소로 함으로써 모집단에서의 값을 추정하는 방법이다.
# 이러한 최소제곱법을 사용하여 증명된 기울기 공식은 다음과 같다.
formula = pilimg.open("C:\\Users\\Msi\\Desktop\\통계 프로그래밍 연습\\회귀분석 기울기 공식.png")
formula
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 3, 2, 5, 4]
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print('mean of x : ', x_mean)
print('mean of y : ', y_mean)
```

```python
x_deviation = x - x_mean
y_deviation = y - y_mean
print('deviation of x : ', x_deviation)
print('deviation of y : ', y_deviation)
```

```python
# 회귀분석의 기울기 구하기
slope = np.sum(x_deviation*y_deviation)/np.sum(np.square(x_deviation))
print('slope : ', slope)
```

```python
# 회귀분석의 절편 구하기
intercept = y_mean - slope * x_mean
print('intercept : ', intercept)
```

```python
# 완성된 회귀식
print('y_hat = %0.1f * x + %0.2f' % (slope, intercept))
```

```python
# 완성된 회귀식에서의 추정값
y_pred = slope * np.array(x) + intercept
print(y_pred)
```

```python
plt.scatter(x, y, c='k')
plt.plot(x, y_pred, c='r')
plt.annotate('y = %0.2f * x + %0.2f' % (slope, intercept), (x[4], y_pred[4]))
plt.show()
```

```python
# 위에서 같이 직접 회귀식을 구할 수도 있지만 아래와 같이 statsmodels.api 또는 scipy.stats 라이브러리를
# 사용한다면 더 쉽고 다양한 통계치들을 구할 수 있다.
# statsmodels.api는 다양한 통계치와 회귀식을 간단한 표로 보여주는 라이브러리이다.
# 여기서 눈여겨 봐야하는 통계치들은 R-squared, coef, p-value이다.
# R-squared(결정계수) 값은 회귀분석에서 회귀식을 생성한 뒤에 회귀식이 얼마나
# 데이터들을 잘 대표하는지를 나타내는 지표이다.
# 밑의 경우에는 R-squared 값이 0.64로 추정된 회귀식이 약 64% 만큼 데이터를 설명하고 있다는 뜻이다.
import statsmodels.api as sm
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
model.summary()
```

```python
# scipy.stats 라이브러리에서도 아래와 같은 방식으로 확인할 수 있다.
import scipy.stats as stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
print("intercept : ", format(intercept, ".2f"))
print("slope : ", format(slope, ".2f"))
print("r_squared : ", format(r_value**2, ".2f"))
print("p_value : ", format(p_value, ".2f"))
print("std_err : ", format(std_err, ".2f"))
```
