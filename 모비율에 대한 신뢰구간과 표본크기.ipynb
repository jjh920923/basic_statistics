```python
def confidence_interval(percent, sample, n):
    p = n / sample
    q = 1 - p
    if percent == 90:
        left = p - 1.65 * math.sqrt((p*q) / sample)
        right = p + 1.65 * math.sqrt((p*q) / sample)
    elif percent == 95:
        left = p - 1.96 * math.sqrt((p*q) / sample)
        right = p + 1.96 * math.sqrt((p*q) / sample)
    elif percent == 99:
        left = p - 2.58 * math.sqrt((p*q) / sample)
        right = p + 2.58 * math.sqrt((p*q) / sample)
    else:
        print("90%, 95%, 99%    .")
    print(" : " + str(format(left, ".3f")) + " <  < " + str(format(right, ".3f")))
```

```python
confidence_interval(90, 500, 60)
```

```python
#         
# error  , percent , std 
def sample_size(error, percent, sample, n):
    p = n / sample
    q = 1 - p
    if percent == 90:
        n = math.ceil((p*q)*math.pow((1.65 / error), 2))
    elif percent == 95:
        n = math.ceil((p*q)*math.pow((1.96 / error), 2))
    elif percent == 99:
        n = math.ceil((p*q)*math.pow((2.58 / error), 2))
    else:
        print("90%, 95%, 99%    .")
    print(" " + str(n) + "  .")
```

```python
sample_size(0.02, 95, 100, 40)
```
